```python
def is_nr(value):
    if type(value) is str:
        try:
            nr = int(value)
            return nr
        except Exception as e:
            print(value)
            return 0
    elif pd.isnull(value):
        return 0
    elif type(value) is float:
        if value.is_integer():
            return value.as_integer_ratio()[0]
        else:
            print("float?")
            print(value)
            return value.as_integer_ratio()[0]
        return value.toint()
    elif type(value) is int:
        return value
    else:
        print(type(value))
        print("what")
        print(value)
```

```python
data["#page_norm"] = data["#Page"].apply(is_nr)
```

```python
data["#line_norm"] = data["#Lines (Cermine)"].apply(is_nr)
```

```python
data["#page_norm"].plot(kind="density")
```

```python
def plot_size_page_ratio(data, dim1="PDF Size (KB)", dim2="#page_norm"):
    if data.shape[0]>1:
        ((data[dim1]/data[dim2])[pd.isnull((data[dim1]/data[dim2]))== False]).plot(kind="density", figsize=(30,8))
```

```python
plot_size_page_ratio(data)
```

```python
data.groupby(["Is Scanned"]).apply(plot_size_page_ratio)
```

```python
data.groupby(["Contains OCR"]).apply(plot_size_page_ratio)
```
